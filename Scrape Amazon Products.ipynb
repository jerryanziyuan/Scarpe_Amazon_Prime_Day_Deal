```python
# import statements
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

 Step 1: Get text from Amazon's airpod search page

```python
# fetch web page
headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64;     x64; rv:66.0) Gecko/20100101 Firefox/66.0",\
           "Accept-Encoding":"gzip, deflate",     "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8",\
           "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"}
#put your url here
r = requests.get('https://www.amazon.com/s?k=airpods&ref=nb_sb_noss', headers=headers)#,proxies=proxies)
```

### Step 4 : Collect price and other data of ALL product listing on the page

```python
product = []
disount_price = []
marked_price = []
rating = []
review = []
for summary in summaries:
    # append name and other data of each product to several lists
    product.append(summary.select_one('h2').text.strip())
    disount_price.append(summary.select_one('span.a-offscreen').text.strip())
    rating.append(summary.select_one('div.a-section.a-spacing-none.a-spacing-top-micro').text.strip().split()[0])
    try:
        review.append(int(summary.select_one('div.a-section.a-spacing-none.a-spacing-top-micro').text.strip().split()[-1].replace(',', '')))
    except: 
        review.append(np.nan)
    try:
        marked_price.append('$'+summary.select_one('span.a-price.a-text-price').text.strip().split('$')[1])
    except: 
        marked_price.append(np.nan)
```

```python
pd.DataFrame(list(zip(product, disount_price,marked_price,rating, review)),\
             columns=['Product', 'disount_price', 'marked_price', 'rating', 'review'])
```

### Create a DataFrame

```python
df_scrape = pd.DataFrame(list(zip(product, disount_price,marked_price,rating, review)),\
             columns=['Product', 'Disount_price', 'Marked_price', 'Rating', 'Review'])
```

### Calculate Discount

```python
df_scrape['Discount'] = pd.to_numeric(df_scrape['Marked_price'].str.replace("$",'') , \
                                      errors='coerce') - pd.to_numeric(df_scrape['Disount_price'].str.replace("$",'') ,\
                                                                       errors='coerce')
```

### Clean Data Format

```python
df_scrape['Rating'] = pd.to_numeric(df_scrape['Rating'], errors='coerce')
df_scrape['Review'] = pd.to_numeric(df_scrape['Review'], errors='coerce').fillna(0, downcast='infer').astype('Int32')
```

```python
df_scrape.sort_values(by = 'Discount',ascending=False)
```
